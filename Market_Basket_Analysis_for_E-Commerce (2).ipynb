# Step 1: Read the dataset

```python
import pandas as pd
file_path = '/content/drive/MyDrive/Market Basket Analysis - Groceries_dataset.csv'
data = pd.read_csv(file_path)
```

# Step 3: Market Basket Analysis
# Data preprocessing - Convert float columns to integers (if needed)

```python
data['Member_number'] = data['Member_number'].astype(int)
data['itemDescription'] = data['itemDescription'].astype(str)  # Convert item description to string (if needed)
```

# Step 4: Data Visualization

# Example: Bar plot of top 100 most purchased items

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
top_items = data['itemDescription'].value_counts().head(20)
plt.figure(figsize=(10, 6))  # Define figure size
sns.barplot(x=top_items.values, y=top_items.index, palette='viridis')
plt.xlabel('Frequency')
plt.ylabel('Item Description')
plt.title('Top 100 Most Purchased Items')
plt.show()
```
